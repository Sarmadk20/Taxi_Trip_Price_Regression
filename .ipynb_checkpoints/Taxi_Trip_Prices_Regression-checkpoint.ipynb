# Taxi_Trip_Price_Regression

# Problem Statement:
You have a dataset containing taxi trip information (such as distance, time of day, day of week, passenger count, traffic conditions, weather, base fare, per kilometer rate, per minute rate, trip duration, and trip price). The dataset contains some missing values (NaN) in columns like "Trip_Distance_km", "Base_Fare", "Per_Km_Rate", etc.
Your task is to analyze this dataset to gain insights, handle missing values, and potentially develop a pricing model or visualization.

# Goal:

- Load the dataset and inspect its contents.
- Clean the dataset by handling missing values.
- Perform exploratory data analysis (EDA) to understand key trends and patterns in taxi trip pricing.
- Visualize the relationships between features and trip price.
- Build a predictive model for taxi trip pricing based on available features.

```python
def percent_missing(df):
    percent_nan = 100*df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

```python
percent_nan = percent_missing(df)
```

```python
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.xticks(rotation=90)
plt.show()
```

```python
num_cols = ['Trip_Distance_km','Passenger_Count','Base_Fare','Per_Km_Rate','Per_Minute_Rate','Trip_Duration_Minutes','Trip_Price']
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
categ_cols = ['Time_of_Day','Day_of_Week','Traffic_Conditions','Weather']
for col in categ_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
percent_nan = percent_missing(df)
```

```python
sns.barplot(x=percent_nan.index,y=percent_nan)
plt.xticks(rotation=90)
plt.show()
```

```python
#  removing outlers from trip-Distancekm

out_rem = df[df["Trip_Distance_km"] > 90].index
df.drop(out_rem, inplace=True)
```

```python
sns.boxplot(x=df["Trip_Distance_km"])
plt.show()
```

```python
df['Day_of_Week'] = df['Day_of_Week'] .map({'Weekday':0,'Weekend':1})
```

```python
encode_col=["Time_of_Day","Traffic_Conditions","Weather"]
file_encode=pd.get_dummies(df,columns=encode_col,dtype=int)
print(file_encode.head())
```

```python
df = file_encode
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
X = df.drop('Trip_Price' ,axis = 1)
y = df['Trip_Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
model = RandomForestRegressor(random_state=101)
```

```python
def run_model(model,X_train,y_train,X_test,y_test):
    
    # Fit Model
    model.fit(X_train,y_train)
    
    # Get Metrics
    preds = model.predict(X_test)


    print("MAE:", mean_absolute_error(y_test, preds))
    print("MSE:", mean_squared_error(y_test, preds))
    print("R^2:", r2_score(y_test, preds))

    plt.figure(figsize=(6,6))
    sns.scatterplot(x=y_test, y=preds,hue=df.loc[y_test.index,'Trip_Price'] ,palette="viridis", alpha=0.7)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title("Predicted vs Actual (Hue = Trip_Price)")
    plt.show()
    
```

```python
run_model(model,X_train,y_train,X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
pred = []
k_vals = [1,5,10]

for n in k_vals:
    model = KNeighborsRegressor(n_neighbors=n)
    run_model(model,X_train,y_train,X_test,y_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

param_grid = {'C':[0.01,0.1,1,5,10,100,1000],'gamma':['auto','scale']}
svr = SVR()
```

```python
grid = GridSearchCV(svr,param_grid)
```

```python
run_model(grid,X_train,y_train,X_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
model = AdaBoostRegressor()
run_model(model,X_train,y_train,X_test,y_test)
```

```python
model = GradientBoostingRegressor()
```

```python
run_model(model,X_train,y_train,X_test,y_test)
```

3. SVM (Support Vector Regression)

4. GradientBoostingRegressor

   
- MAE: 5.38MSE: 66.61R: 0.90
- Interpretation: This model outperforms the rest, achieving the lowest errors and highest R. Gradient boosting is well-suited for structured regression tasks and maximizes predictive accuracy, explaining 90% of variance.


5. AdaBoostRegressor
   
- MAE: 9.83MSE: 143.99R: 0.77
- Interpretation: Decent performance but not as strong as Random Forest or Gradient Boosting. It is sensitive to noisy data and might struggle if the weak learners aren't capturing enough complexity.

# Final Insights

- Tree-based ensemble models (Random Forest, Gradient Boosting) provide the best predictions for taxi trip pricing in your data. They are robust to outliers and can capture non-linear interactions and feature importance.
- GradientBoostingRegressor is your top-performing model, with the lowest mean error and highest explanatory power.
- KNN is least suitable for this regression problem, likely due to high dimensionality and lack of feature scaling or poor data granularity.
- SVM offers intermediate performance, benefitting from kernel tricks but needing more tuning.
- AdaBoost is effective, but less so than Random Forest/Gradient Boosting.
